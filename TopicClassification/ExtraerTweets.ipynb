```python
import pickle
import time
import tweepy

#CÃ³digo para la api de twitter, con keys y tokens

consumer_key = ''
consumer_secret = ''
access_key = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth) 
```

```python

#FunciÃ³n que define el uso del cursor para extracciÃ³n de tweets, si todo va bien va de tweet en tweet, si hay algÃºn error 
#(como llegar al lÃ­mite de descargad de la api) hace un sleep del tiempo suficiente para poder hacer la extracciÃ³n de tweets otra vez,

def limitHandled(cursor, users):
  while True:
    try:
        yield cursor.next()
    except (tweepy.errors.TweepyException, StopIteration):
        if len(users)>18000:
            print("START SLEEPING!")
            time.sleep(15 * 60)
            print("END SLEEPING!")

#Los tÃ³picos elegidos son polÃ­tica, viaje, electrÃ³nica y coches y sus variaciones. Esto se ha elegido asÃ­ porque investigando sobre el tema online no existen 
#demasiada informaciÃ³n sobre multilabel classifcation como en este caso, pero uno de los papers mecionaba que habÃ­a usado estos temas con resultados satisfactorios.

topics = ['politics','politics and travel','politics and car','electronics and car','electronics','car','travel','travel and electronics','travel and car']

#Al ejecutar el cÃ³digo llama a la funciÃ³n anterior y a cursor, que permite seleccionar los tweeets que se quieren (en este caso 3000, aunque se usen solo 2500 para 
#evitar fallos en la api), el tÃ³pico de los mismos y el idioma, de esos tweets se extrae el contenido y el usuario y se almacena en un archivo pickle.
if __name__ == '__main__':
  for topic in topics:
    users = {}
    for tweet in limitHandled(tweepy.Cursor(api.search_tweets, q=(topics[0]), count=100, lang='en').items(3000), users):
      print(len(users))
      print(tweet.user.name)
      print(tweet.text)
      if tweet.user.id in users:
        users[tweet.user.id].append(tweet.text)  
      else:
        users[tweet.user.id] = []
        users[tweet.user.id].append(tweet.text)  
      if len(users)==2500:
        break
    fileName = topic + '_train_data9.data'
    print (fileName)
    userTweets = open(fileName,'ab+')
    pickle.dump(users,userTweets)
    userTweets.close() 
```
